Treating Null Values

```python
df.isnull().sum()
```

```python
# Sort properly
df = df.sort_values(['Country', 'Year'])

# Forward/backward fill within country
df = df.groupby('Country').apply(lambda x: x.ffill().bfill())

# Fill any remaining numeric nulls with median
df = df.fillna(df.median(numeric_only=True))

# Fill categorical separately
df['Vaccine_stock_status'] = df['Vaccine_stock_status'].fillna('Unknown')
```

```python
numeric_cols = [
    'Suspected_measles_cases',
    'MCV2_coverage (%)',
    'Health_expenditure_per_capita (US$)',
    'Dropout_rate (MCV1?MCV2)',
    'Measles_incidence_rate (per million)'
]
```

```python
df = df.reset_index(drop=True)
```

```python
# Remove commas and convert to numeric
for col in numeric_cols:
    df[col] = (
        df[col]
        .astype(str)                # ensure string
        .str.replace(',', '', regex=False)  # remove commas
        .replace('nan', np.nan)     # fix string 'nan'
    )

    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python

def clean_numeric(col):
    return (
        col.astype(str)
           .str.replace(',', '', regex=False)
           .str.replace('%', '', regex=False)
           .str.replace('$', '', regex=False)
           .str.strip()
           .replace(['', 'nan', 'None'], np.nan)
    )

# Apply to suspected numeric columns
for col in numeric_cols:
    df[col] = pd.to_numeric(clean_numeric(df[col]), errors='coerce')
```

```python
df.info()
```

```python
df['Population'] = df['Population'].str.replace(',', '', regex=False)
df['Population'] = pd.to_numeric(df['Population'], errors='coerce')
```

```python
# df['Reported_measles_cases'].unique()
df['Reported_measles_cases'] = df['Reported_measles_cases'].str.replace(',', '', regex=False)
df['Reported_measles_cases'] = pd.to_numeric(df['Reported_measles_cases'], errors='coerce')
```

```python
# df['Measles_deaths'].unique()
df['Measles_deaths'] = df['Measles_deaths'].str.replace(',', '', regex=False)
df['Measles_deaths'] = pd.to_numeric(df['Measles_deaths'], errors='coerce')
```

```python
df['MCV1_coverage (%)'].unique()
df['MCV1_coverage (%)'] = df['MCV1_coverage (%)'].str.replace(',', '', regex=False)
df['MCV1_coverage (%)'] = pd.to_numeric(df['MCV1_coverage (%)'], errors='coerce')
```

```python
df['DTP3_coverage (%)'].unique()
df['DTP3_coverage (%)'] = df['DTP3_coverage (%)'].str.replace(',', '', regex=False)
df['DTP3_coverage (%)'] = pd.to_numeric(df['DTP3_coverage (%)'], errors='coerce')
```

```python
# df['Population_density (people/kmÂ²)'].unique()
df['Population_density (people/kmÂ²)'] = df['Population_density (people/kmÂ²)'].str.replace(',', '', regex=False)
df['Population_density (people/kmÂ²)'] = pd.to_numeric(df['Population_density (people/kmÂ²)'], errors='coerce')
```

```python
features_to = ['Birth_rate (per 1,000)','Proportion_under5 (%)','GDP_per_capita (US$)','Hospital_beds_per_1000','Vaccine_stock_status','Air_travel_volume (million passengers)','Internet_penetration (%)','Routine_immunization_dropout (%)','Avg_Annual_Temperature','Total_Annual_Rainfall','Avg_Annual_Humidity','Rainy_Season_Length','Temp_Seasonality','Dry_Season_Length','Extreme_Rain_Days']
```

```python
for col in features_to:
    # Skip if the column is categorical
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.replace(',', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python

categorical_cols = ['Country', 'Vaccine_stock_status', 'Outbreak_alert', 'Outbreak_Label']
numeric_cols = [
    'Population', 'Population_density (people/kmÂ²)',
    'Reported_measles_cases','Suspected_measles_cases','Measles_deaths',
    'MCV1_coverage (%)','MCV2_coverage (%)','DTP3_coverage (%)',
    'Birth_rate (per 1,000)','Proportion_under5 (%)',
    'GDP_per_capita (US$)','Health_expenditure_per_capita (US$)',
    'Hospital_beds_per_1000','Physicians_per_10000',
    'Air_travel_volume (million passengers)','Internet_penetration (%)',
    'Routine_immunization_dropout (%)','Dropout_rate (MCV1?MCV2)',
    'Measles_incidence_rate (per million)',
    'Avg_Annual_Temperature','Total_Annual_Rainfall','Avg_Annual_Humidity',
    'Rainy_Season_Length','Temp_Seasonality','Dry_Season_Length','Extreme_Rain_Days'
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = df[col].astype(str) \
                        .str.replace(',', '', regex=False) \
                        .str.replace('%', '', regex=False) \
                        .str.replace('$', '', regex=False) \
                        .str.strip() \
                        .replace(['', 'nan', 'None'], np.nan)
        df[col] = pd.to_numeric(df[col], errors='coerce')


for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')



for col in numeric_cols:
    if col in df.columns and df[col].isnull().any():
        df[col] = df.groupby('Country')[col].transform(lambda x: x.ffill().bfill())


for col in numeric_cols:
    if col in df.columns and df[col].isnull().any():
        df[col] = df.groupby('Country')[col].transform(lambda x: x.fillna(x.median()))

for col in numeric_cols:
    if col in df.columns and df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())


print(df.isnull().sum())
```
